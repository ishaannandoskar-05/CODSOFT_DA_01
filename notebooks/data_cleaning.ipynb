```python
import pandas as pd
import numpy as np

# Load dataset (relative path assumes notebook is run from notebooks/ folder)
df = pd.read_csv('../data/raw/titanic.csv')
df.head()
```

```python
print("Shape:", df.shape)
df.info()
```

```python
df.describe()
```

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
```

```python
print("Duplicate rows:", df.duplicated().sum())
df[df.duplicated()]
```

```python
print("Unique Sex values:", df['Sex'].unique())
print("Unique Embarked values:", df['Embarked'].unique())
print("Unique Pclass values:", sorted(df['Pclass'].unique()))
print("Age range:", df['Age'].min(), "-", df['Age'].max())
print("Fare range:", df['Fare'].min(), "-", df['Fare'].max())
```

```python
original_shape = df.shape

# Fill missing Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill missing Embarked with mode
mode_embarked = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(mode_embarked)

# Cabin is 77% missing -> convert to a binary flag instead of imputing or keeping raw
df['Has_Cabin'] = df['Cabin'].notna().astype(int)
df.drop(columns=['Cabin'], inplace=True)

# Remove duplicate rows
df = df.drop_duplicates()

# Correct data types
df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
df['Sex'] = df['Sex'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')

# Strip whitespace from text columns
for col in ['Sex', 'Embarked']:
    df[col] = df[col].astype(str).str.strip()

df.reset_index(drop=True, inplace=True)

print("Shape before:", original_shape, "-> after:", df.shape)
```

```python
print("Remaining missing values:")
print(df.isnull().sum())
print("\nFinal dtypes:")
print(df.dtypes)
df.head()
```

```python
df.to_csv('../data/processed/titanic_cleaned.csv', index=False)
print("Saved to ../data/processed/titanic_cleaned.csv")
```
